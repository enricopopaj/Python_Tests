```python
#df = pd.read_csv('/Applications/Hobbys/Finanzen/Test/Umsatzauskunft_KtoNr0698877125_30-04-2020_20-14-04.csv', skiprows=13, sep=';')#, encoding='cp1252')#, encoding='cp1252', skiprows=15)
#df = pd.read_csv('/Applications/Hobbys/Finanzen/Test/Umsatzauskunft_KtoNr0698877125_30-04-2020_20-14-04.xlsx', encoding='utf-8', error_bad_lines=False)
#df = pd.read_xml('/Applications/Hobbys/Finanzen/Test/Umsatzauskunft_KtoNr0698877125_30-04-2020_23-29-44.xml')

# Standard Process
file_Path = '/Users/enricopopaj/Desktop/'

csv_Files = [file for file in listdir(file_Path) if isfile(join(file_Path, file)) if 'Umsatzauskunft' in file]


df = pd.read_csv(file_Path + csv_Files[0], skiprows=13, sep=';')

df
```

```python
df = df.rename(columns={'Empf‰nger': 'Empfänger', 'Betrag (Ä)': 'Betrag', 'Saldo (Ä)': 'Saldo'})
df['Betrag'] = df['Betrag'].str.replace(' Ä', '')
df['Betrag'] = df['Betrag'].str.replace('-', '')
df['Saldo'] = df['Saldo'].str.replace(' Ä', '')
df['Saldo'] = df['Saldo'].str.replace('-', '')
df
```

```python
uArt = []
empf = []
for index, row in df.iterrows():
    #print(row[2])
    
    bargeld = ['Bargeldausz. im Handel', 'Bargeldausz. Geldautomat', 'Auszahlung']
    einkauf = ['LIDL', 'REWE', 'KAUFLAND', 'ALDI', 'EDEKA']
    
    
    if row[2] in bargeld:
        uArt.append('Auszahlung')
    elif (row[2] != 'Gehalt/Rente'):
        uArt.append('Zahlung')
    else:
        uArt.append(row[2])
        
    if any(x in str(row[5]) for x in einkauf):
        empf.append('Lebensmittel')
    elif 'AMAZON' in str(row[5]):
        empf.append('AMAZON')
    elif 'PayPal' in str(row[5]):
        empf.append('PayPal')
    else: 
        empf.append(str(row[5]))
    
    
    

print(uArt)
print(empf)

df['Umsatzart'] = uArt
df['Empfänger'] = empf
df
```
