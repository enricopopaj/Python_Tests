```python
import requests
import json
import csv
import re 
from bs4 import BeautifulSoup
import statsmodels.formula.api as smf
import lxml.html as lh
from statsmodels.graphics.regressionplots import abline_plot
import pandas as pd
```

```python
# base_url = "https://www.bernhard-gaul.de/wissen/staatenerde.php#uebneu"
base_url = 'https://www.goerres-pellarin.de/2012/01/laender-hauptstaedte-kontinente/'
page = requests.get(base_url)
code = BeautifulSoup(page.content, 'html.parser')

print(code.prettify())
```

```python
table = code.find('div', class_='post-336')

land = []
hauptstadt = []
kontinent = []
alle = []

# Auflisten aller Einträge
i = 1

for value in code.find_all('td'):
    if i >= 4:
        value = str(value)
        value = re.search('>(.*)</td>', value).group(1)
        
        alle.append(value)
        
    i += 1

l = 0
h = 1
k = 2

# Einträge nach Land, Hauptstadt und Kontinent zuordnen

while k <= len(alle):

    # Land
    land.append(alle[l])
    print('Land ', alle[l])
    l += 3

    # Hauptstadt
    hauptstadt.append(alle[h])
    print('Hauptstadt ', alle[h])
    h += 3

    # Kontinent
    if alle[k].startswith('Amerika/'):
        value = alle[k].split('/')[1]
        kontinent.append(value)
    else:
        kontinent.append(alle[k])
    print('Kontinent ', alle[k])
    k += 3
```

```python
df = pd.DataFrame({'Kontinent': kontinent, 'Land': land, 'Hauptstadt': hauptstadt})
df.head(60)
```
